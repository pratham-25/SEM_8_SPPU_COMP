```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.metrics import mean_absolute_error, mean_squared_error

from keras.models import Sequential
from keras.layers import Dense
```

```python
df = pd.read_csv('BostonHousing.csv')
df
```

```python
df.shape
```

```python
df.dtypes
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
X = df.iloc[:,df.columns != 'medv']
X
```

```python
y = df.iloc[:,df.columns == 'medv']
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33,random_state=42) 
```

```python
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_train
```

```python
X_test = scaler.transform(X_test)
X_test
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), annot=True,cmap='coolwarm')
```

```python
model = Sequential()
```

```python
model.add(Dense(units = 128, activation = 'relu', input_shape = (X_train.shape[1],)))
model.add(Dense(units = 64, activation = 'relu'))
model.add(Dense(units = 1, activation = 'linear'))
model.compile(optimizer='adam', loss = 'mean_squared_error')
model.summary()
```

```python
history = model.fit(X_train, y_train, epochs=250, batch_size = 32, validation_split=0.05)
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_test.iloc[:,0]
```

```python
d = pd.DataFrame({'y_test': y_test.iloc[:,0], 'y_pred': y_pred[:,0]})
d
```

```python
mse = mean_squared_error(y_test,y_pred)
mae = mean_absolute_error(y_test,y_pred)
print(f'Mean Squared Error: {mse}')
print(f'Mean Absolute Error: {mae}')
```

```python
plt.plot(history.history['loss'], label='Training')
plt.plot(history.history['val_loss'], label='Validation')
plt.legend()
plt.show()
```

```python
plt.plot(d['y_test'].head(),'ro',label = 'Actual')
plt.plot(d['y_pred'].head(),'bo', label = 'Predicted')
plt.legend()
plt.show()
```
